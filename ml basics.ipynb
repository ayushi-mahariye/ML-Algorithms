## Types of Machine Learning
#### 1. Supervised Machine Learning :
 ##### Supervised machine learning learns patterns and relationships between input and output data. It is defined by its use of labeled data
 ##### There are two types of supervised learning algorithms:
1. classification
2. Regression



## Classification
1. Classification is a supervised machine learning method where the model try to predict the correct output based on the given input Data
2. In classification the model is fully trained on the training dataset aand then evaluated on test data set  before to make prediction on new data
3. For Example:  Email Spam Classifier
### Mathematical Formula : 
y=f(x) x: input variable, y: categorical output

## There are two types of classifications: 
1. Binary Classifier: If the classification problem has only two possible outcomes, then it is called as Binary Classifier.
Examples: YES or NO, MALE or FEMALE, SPAM or NOT SPAM, CAT or DOG, etc.
2. Multi-class Classifier: If a classification problem has more than two outcomes, then it is called as Multi-class Classifier.
Example: Classifications of types of crops, Classification of types of music.


## Regression
Regression is a type of supervised machine learning where algorithms learn from the data to predict continuous values such as sales, salary, weight, or temperature. For example:

For Example: A dataset containing features of the house such as lot size, number of bedrooms, number of baths, neighborhood, etc. and the price of the house, a Regression algorithm can be trained to learn the relationship between the features and the price of the house.



## There are many machine learning algorithms that can be used for classification tasks. Some of them are:

3. K Nearest Neighbor Classifier
4. Random Forest Classifier
5. Neural Networks

## There are many machine learning algorithms that can be used for regression tasks. Some of them are:
